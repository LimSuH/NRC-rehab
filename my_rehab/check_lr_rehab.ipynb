```python
import os
import numpy as np
import pandas as pd
import csv
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
with open("/home/neuronS1/NRC/my_rehab/caseRecord/exp4/log_2025-03-17 16:47:12_18.pkl", 'rb') as f:
        log = pkl.load(f)
```

```python
mat = log
```

```python
matnp = np.array(mat['error'])
```

```python
err=[]
val_err=[]
```

```python
for i in mat['error']:
    err.append(max(i))

for i in mat['val_error']:
    val_err.append(max(i))
```

```python
FIG, AXES = plt.subplots(1,2,figsize=(12,5))

ax1 = AXES[0].twiny()
ax2 = AXES[1].twiny()


AXES[0].plot(err, color='blue', label='train')
ax1.plot(val_err, color='red', label='validation')

AXES[1].plot(mat['loss'], color='blue')
ax2.plot(mat['val_loss'], color='red')

AXES[0].set_title('error+-')
AXES[1].set_title('loss')

FIG.legend()
FIG.supxlabel('Epoch')
```
